```python
ax=sns.countplot(x='finalWorth',data=df)
sns.set(rc={'figure.figsize':(100,50)})
```

```python
ax=sns.countplot(x='finalWorth',data=df)
sns.set(rc={'figure.figsize':(100,50)})
for value in ax.containers:
    ax.bar_label(value)
```

```python
ax=sns.countplot(x='age',data=df)
sns.set(rc={'figure.figsize':(100,50)})
for value in ax.containers:
    ax.bar_label(value)
```

```python
ax=sns.countplot(x='Rank',data=df)
sns.set(rc={'figure.figsize':(50,0)})
for value in ax.containers:
    ax.bar_label(value)
```

```python
df.groupby(['Rank'],as_index=False)['finalWorth'].sum().sort_values(by='finalWorth',ascending=False)
```

```python
df.groupby(['Rank'],as_index=False)['finalWorth'].sum().sort_values(by='finalWorth',ascending=False)
```

```python
df.sort_values(by='finalWorth',ascending=False)
```

```python
rank=df.groupby(['Rank'],as_index=False)['finalWorth'].sum().sort_values(by='finalWorth',ascending=False)
sns.set(rc={'figure.figsize':(20,10)})
sns.lineplot(x='Rank',y='finalWorth',data=rank)
```
